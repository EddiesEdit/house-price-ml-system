```python
class DataLoader:

    def __init__(self, filepath):
        self.filepath = filepath
        self.data = None

    def load_data(self):
        self.data = pd.read_csv(self.filepath)
        print("Data loaded Successfully")
        return self.data

    def preview_data(self, rows=5):
        """Preview First Five Rows"""
        return self.data.head(rows)

    def data_info(self):
        """Display Data Info"""
        return self.data.info()

    def describe_data(self):
        """Statistical Summary"""
        return self.data.describe()


    def check_missing(self):
        return self.data.isnull().sum()

    def check_unique_counts(self):
        return self.data.nunique()

    def show_unique_values(self):
        for col in self.data.columns:
            print(f"\nColumn: {col}")
            print(self.data[col].unique())

    def value_counts(self, column):
        return self.data[column].value_counts()
        
```

### Data Cleaning

```python
class DataCleaner:
    
    def __init__(self, data):
        self.data = data.copy()

    def __repr__(self):
        return repr(self.data)

    def fill_missing(self):
        """numeric columns"""
        numeric_cols = self.data.select_dtypes(include=['number']).columns

        """categorical columns"""
        cat_cols = self.data.select_dtypes(include=['object']).columns

        """fill numeric columns"""
        self.data[numeric_cols] = self.data[numeric_cols].fillna(
            self.data[numeric_cols].mean()
        )

        """fill categorical columns with mode"""
        for col in cat_cols:
            self.data[col] = self.data[col].fillna(
                self.data[col].mode()[0]
            )

        print("Missing values handled for numeric and categorical columns")
        
        return self.data
            
```

### Feature Engineering

```python
class FeatureEngineer:

    def __init__(self, data):
        self.data = data.copy()

    def __repr__(self):
        return repr(self.data)

    def encode_binary(self):

        binary_cols = [
            "mainroad",
            "guestroom",
            "basement",
            "hotwaterheating",
            "airconditioning",
            "prefarea"
        ]

        for col in binary_cols:
            self.data[col] = self.data[col].map({
                "yes": 1,
                "no": 0
            })

        print("✅ Binary columns encoded")
        return self.data


    def encode_furnishing(self):

        self.data = pd.get_dummies(
            self.data,
            columns=["furnishingstatus"],
            drop_first=True
        )

        print("✅ Furnishing status encoded")
        return self.data


    def split_features_target(self):

        X = self.data.drop("price", axis=1)
        y = self.data["price"]

        print("✅ Features and target created")

        return X, y
```

### Preprocessing 

```python
from sklearn.preprocessing import StandardScaler

class FeatureScaler:

    def __init__(self):
        self.scaler = StandardScaler()

    def scale(self, X_train, X_test):

        # Fit only on training data
        X_train_scaled = self.scaler.fit_transform(X_train)

        # Transform test data
        X_test_scaled = self.scaler.transform(X_test)

        # Convert back to DataFrame
        X_train_scaled = pd.DataFrame(
            X_train_scaled,
            columns=X_train.columns
        )

        X_test_scaled = pd.DataFrame(
            X_test_scaled,
            columns=X_test.columns
        )

        print("✅ Training and testing features scaled")

        return X_train_scaled, X_test_scaled
```

### Feature Importance

```python

class FeatureImportance:

    def __init__(self, model, feature_names):
        self.model = model
        self.feature_names = feature_names

    def plot_importance(self):

        importance = self.model.coef_

        importance_df = pd.DataFrame({
            "feature": self.feature_names,
            "importance": importance
        }).sort_values(by="importance", ascending=False)

        plt.figure(figsize=(8,6))

        sns.barplot(
            x="importance",
            y="feature
            data=importance_df
        )

        plt.title("Feature Importance (Linear Regression Coefficients)")
        plt.show()

        return importance_df
```

```python
from sklearn.model_selection import cross_val_score
import numpy as np

class CrossValidator:

    def __init__(self, model, cv=5):
        self.model = model
        self.cv = cv

    def evaluate(self, X, y):

        scores = cross_val_score(
            self.model,
            X,
            y,
            scoring="r2",
            cv=self.cv
        )

        print("Cross Validation Results")
        print("------------------------")
        print(f"Scores: {scores}")
        print(f"Mean R²: {scores.mean()}")
        print(f"Std: {scores.std()}")

        return scores
```

### Model Comparison

```python

from sklearn.model_selection import cross_val_score
from sklearn.metrics import r2_score


class ModelComparison:

    def __init__(self, models, cv=5):
        """
        models : dictionary of models
        cv : number of folds for cross validation
        """
        self.models = models
        self.cv = cv
        self.results = []

    def train_and_compare(self, X_train, X_test, y_train, y_test):

        for name, model in self.models.items():

            # Train model
            model.fit(X_train, y_train)

            # Predict
            predictions = model.predict(X_test)

            # Test R2
            r2 = r2_score(y_test, predictions)

            # Cross validation
            cv_scores = cross_val_score(
                model,
                X_train,
                y_train,
                cv=self.cv,
                scoring="r2"
            )

            result = {
                "Model": name,
                "Test_R2": r2,
                "CV_Mean": cv_scores.mean(),
                "CV_STD": cv_scores.std()
            }

            self.results.append(result)

        self.results_df = pd.DataFrame(self.results).sort_values(
            by="CV_Mean",
            ascending=False
        )

        return self.results_df

    def plot_comparison(self):

        plt.figure(figsize=(8,6))

        sns.barplot(
            data=self.results_df,
            x="CV_Mean",
            y="Model"
        )

        plt.title("Model Performance Comparison (Cross Validation R²)")
        plt.xlabel("Mean CV R²")
        plt.ylabel("Model")

        plt.show()
```

```python
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor

models = {

    "Linear Regression": LinearRegression(),

    "Ridge Regression": Ridge(),

    "Lasso Regression": Lasso(),

    "Decision Tree": DecisionTreeRegressor(),

    "Random Forest": RandomForestRegressor()

}
```

```python
comparison = ModelComparison(models)

results = comparison.train_and_compare(
    X_train_scaled,
    X_test_scaled,
    y_train,
    y_test
)

results
```
